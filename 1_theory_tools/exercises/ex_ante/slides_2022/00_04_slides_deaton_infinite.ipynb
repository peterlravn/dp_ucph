# Exercise 7: Gaussian Shock

Consider again the model from exercise 6, but with an infinite time horizon and interest, $R=1$:
$$\begin{eqnarray*}
V(W_{t}) & = & \max_{C_{t}}\left\{ \frac{C_{t}^{1-\rho}}{1-\rho}+\beta \int_{-\infty}^{\infty} V(W_{t+1})g(y_{t+1})dy_{t+1}\right\} \\
 & \text{s.t.}\\
W_{t+1} & = & R(W_{t}-C_{t})+y_{t+1}\\
y_{t+1}  & = & \exp{(\xi_{t+1})}\\
\xi_{t+1} & \sim  & \mathcal{N}(0,\sigma_{\xi}^{2})\\
A_{t}=W_{t}-C_{t} & \geq & 0
\end{eqnarray*}$$

for $T\to \infty$, $\beta=0.90$, $\rho=0.5$, $W=10$ and $\sigma_{\xi}=0.2$.

remember from last time, we approximate the integral with a discrete sum using quadrature:

$$\begin{eqnarray*}
V(W_{t}) & = & \max_{C_{t}}\left\{ \frac{C_{t}^{1-\rho}}{1-\rho}+\beta \int_{-\infty}^{\infty} V(W_{t+1})g(y_{t+1})dy_{t+1}\right\} \approx \max_{C_{t}}\left\{ \frac{C_{t}^{1-\rho}}{1-\rho}+\beta \sum_{k=1}^K w_k V(\underbrace{W_{t+1}}_{R(W_{t}-C_{t})+y_{t+1}^k})\right\} 
\end{eqnarray*}$$

How do the quadrature points and weights look?




## How do we solve the infinite model?

We can use value function iteration (VFI).

The idea is to find the function $V(W_t)$ that satifies:

$$\begin{eqnarray*}
V^{n+1}(W_{t}) & \approx \max_{C_{t}}\left\{ \frac{C_{t}^{1-\rho}}{1-\rho}+\beta \sum_{k=1}^K w_k V^{n}(W_{t+1})\right\} 
\end{eqnarray*}$$

Notice, normally our unknown is not a function but a single point, i.e. $x=-x+2\implies x=1$.




  

Thus, we need to find a function that satisfies the equation:

$$\begin{eqnarray*}
V^{n+1}(W_{t}) & \approx \max_{C_{t}}\left\{ \frac{C_{t}^{1-\rho}}{1-\rho}+\beta \sum_{k=1}^K w_k V^{n}(W_{t+1})\right\} 
\end{eqnarray*}$$




stop when $n=200$ or $|V^{n+1}-V^{n}|$ is small.
