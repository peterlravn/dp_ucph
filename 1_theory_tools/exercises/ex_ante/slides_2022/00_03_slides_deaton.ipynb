# Exercise 6: Gaussian Shock

Consider the model
$$\begin{eqnarray*}
V_{t}(W_{t}) & = & \max_{C_{t}}\left\{ \frac{C_{t}^{1-\rho}}{1-\rho}+\beta \int_{-\infty}^{\infty} V_{t+1}(W_{t+1})g(y_{t+1})dy_{t+1}\right\} \\
 & \text{s.t.}\\
W_{t+1} & = & R(W_{t}-C_{t})+y_{t+1}\\
y_{t+1}  & = & \exp{(\xi_{t+1})}\\
\xi_{t+1} & \sim  & \mathcal{N}(0,\sigma_{\xi}^{2})\\
A_{t}=W_{t}-C_{t} & \geq & 0
\end{eqnarray*}$$

for $T=10$, $\beta=0.98$, $\rho=0.5$, $R=1.0/\beta$, $W_1=10$ and $\sigma_{\xi}=0.2$.

What is actually new in this fancy pancy Deaton model? Not much!

1) New utility function: $u(C_t)=\frac{C_{t}^{1-\rho}}{1-\rho}$



2) Earn interst on $W_t-C_t$: $R=1/\beta \approx 1.02$

3) Last time the shocks were uniformly distributed: 





3) The shocks are now log-normal: $\begin{eqnarray*}
y_{t+1}  & = & \exp{(\xi_{t+1})},
\xi_{t+1} & \sim  & \mathcal{N}(0,\sigma_{\xi}^{2})
\end{eqnarray*}$






We can approximate the integral using quadrature:

$$\begin{eqnarray*}
V_{t}(W_{t}) & = & \max_{C_{t}}\left\{ \frac{C_{t}^{1-\rho}}{1-\rho}+\beta \int_{-\infty}^{\infty} V_{t+1}(W_{t+1})g(y_{t+1})dy_{t+1}\right\} \approx \max_{C_{t}}\left\{ \frac{C_{t}^{1-\rho}}{1-\rho}+\beta \sum_{k=1}^K w_k V_{t+1}(\underbrace{W_{t+1}}_{R(W_{t}-C_{t})+y_{t+1}^k})\right\} 
\end{eqnarray*}$$

## Next up: How do we check the accuracy of our model?

We are doing all kinds of approximations!

* Grid for resources and grid for consumption

* Numerical integration using quadrature

Maybe the solution is inaccurate?


  

Remember the (determinstic) Euler equation from microeconomics (or your horrible macro course):

$$\underbrace{u'(C_{it}^*)}_{\text{marginal utility today}} = \underbrace{\beta R u'(C_{it+1}^*)}_{\text{marginal utility tomorrow}}$$

The marginal utility today must be equal to the marginal utility tomorrow.

In the stochastic case the Euler equation becomes:
    
$$\underbrace{u'(C_{it}^*)}_{\text{marginal utility today}} = \underbrace{\beta R \sum_{k=1}^K w_k u'(R(W_{it}-C_{it}^*)+y_{it+1}^k)}_{\text{Expected marginal utility tomorrow}}$$

The marginal utility today must be equal to the expected marginal utility tomorrow.

We can use this to evaluate the precision of the solution using the absolute value of the "Euler-residual":

$$\left|\epsilon_{it} \right|=\left|u'(C_{it}^*) -\beta R \sum_{k=1}^K w_k u'(R(W_{it}-C_{it}^*)+y_{it+1}^k) \right|$$

Average over individuals and time (only for the interior solution):

$$\begin{eqnarray*}
 \frac{1}{\sum_{i=1}^{N}\sum_{t=1}^{T}  \mathbb{1}_{\{0<C_1<W_{it}\}}}\sum_{i=1}^{N}\sum_{t=1}^{T} \mathbb{1}_{\{0<C_1<W_{it}\}} |\varepsilon_{it}|
 \end{eqnarray*}$$
