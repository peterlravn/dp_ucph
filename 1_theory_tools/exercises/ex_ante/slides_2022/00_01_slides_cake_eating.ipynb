## Exercise 1: The Simplest Consumption Model (Cake Eating Problem):
    
Consider the simplest consumption-saving model:

$$\begin{aligned}
    V_1^*(W) &= \max_{c_1,c_2,\dots,c_T} \{ \sqrt{c_1} +\beta \sqrt{c_2}+\beta^2\sqrt{c_3}+\cdots+\beta^T\sqrt{c_T} \} \\
    & \text{s.t.} \\
    W &= c_1 + c_2+\cdots+c_T \\

\end{aligned}$$
for $T=3$, $\beta =0.9$ and $W=5$.

## Exercise 2: The Infinite Horizon Consumption Model
    
Consider again the simplest consumption-saving model but with an infinite time horizon:

$$\begin{aligned}
    V_1^*(W_1) &= \max_{c_1,c_2,\dots} \{ \sqrt{c_1} +\beta \sqrt{c_2}+\beta^2\sqrt{c_3}+\cdots \} \\
    & \text{s.t.} \\
    W_{t+1} & = W_{t}-c_{t} \\
    c_{t} & \leq  W_{t} , \,\forall t\\
    c_{t} & \in  \mathbb{R}_+, \,\forall t
\end{aligned}$$
for $\beta =0.9$ and $W_1=5$.



Notice as before, we can rewrite the first equation as a Bellman equation:

$$\begin{aligned}
    V^*(W_t) &=\max_{c_t}\{\sqrt{c_t} +\beta V^*(\underbrace{W_{t}-c_t}_{W_{t+1}}) \}, \forall t
\end{aligned}$$

Backward induction is no longer feasible for $t \to \infty$. Instead we can use value function iteration (VFI). How does this work?

Start with an initial guess on the value function, for instance $V^*_1=\textbf{0}$:

$$\begin{aligned}
    V^*_2(W) &=\max_{c}\{\sqrt{c} +\beta \underbrace{V_1^*(W-c)}_0 \}=\max_{c}\sqrt{c} 
\end{aligned}$$

First iteration is the same as in backward induction. We will spent remaining resources!

     


Keep on iterating: 

$$\begin{aligned}
\begin{matrix}
           V^*_3(W) &=\max_{c} \{ \sqrt{c} +\beta V_2^*(W-c) \} \\
           V^*_4(W) &=\max_{c} \{ \sqrt{c} +\beta V_3^*(W-c) \} \\
            \vdots &  \vdots\\
           V^*_{N+1}(W) &=\max_{c} \{ \sqrt{c} +\beta V^*_{N}(W-c) \}
\end{matrix}
\end{aligned}$$   

This is exactly the same as backward iteration back in time. We just keep on doing it until the value function barely changes.
